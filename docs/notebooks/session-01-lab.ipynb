## Part 2: The AI vs ML vs DL Hierarchy

Visualize the relationship between AI, ML, and Deep Learning:

```python
hierarchy = {
    'AI (Artificial Intelligence)': {
        'definition': 'Any technique that enables machines to mimic human intelligence',
        'examples': ['Chess engines', 'Spam filters', 'ChatGPT', 'Face recognition'],
        'sub_fields': {
            'Machine Learning': {
                'definition': 'AI that learns from data',
                'examples': ['Spam filter', 'Netflix recommendations'],
                'sub_fields': {
                    'Deep Learning': {
                        'definition': 'ML using neural networks with many layers',
                        'examples': ['Image recognition', 'GPT models', 'Voice assistants'],
                        'sub_fields': {
                            'Generative AI': {
                                'definition': 'DL that creates new content',
                                'examples': ['ChatGPT', 'DALL-E', 'Stable Diffusion']
                            }
                        }
                    }
                }
            }
        }
    }
}

def print_hierarchy(d, indent=0):
    for name, info in d.items():
        print(' ' * indent + f'📦 {name}')
        print(' ' * indent + f'   Definition: {info["definition"]}')
        print(' ' * indent + f'   Examples: {', '.join(info["examples"])}')
        if 'sub_fields' in info:
            print_hierarchy(info['sub_fields'], indent + 4)

print_hierarchy(hierarchy)
```

```python
import anthropic
import os

# Set your API key
# In Colab: use the Secrets panel (key icon) to store ANTHROPIC_API_KEY
os.environ.setdefault('ANTHROPIC_API_KEY', 'your-key-here')

client = anthropic.Anthropic()

def ask(prompt, system='', max_tokens=300):
    resp = client.messages.create(
        model='claude-haiku-4-5-20251001',
        max_tokens=max_tokens,
        system=system,
        messages=[{'role': 'user', 'content': prompt}]
    )
    return resp.content[0].text

print('✅ Client ready')
```

## Part 1: Bad Prompt → Good Prompt

Rewrite each bad prompt using the anatomy: Role + Task + Context + Constraints + Format

```python
bad_prompts = [
    'Write something about climate change',
    'Summarize this',
    'Help me with my email',
    'What is Python?',
]

# Example good rewrite:
good_example = """You are a science communicator writing for a general audience.
Task: Write a 3-bullet summary of the most urgent climate change risks.
Context: Audience is high school students with no science background.
Constraints: Each bullet must be under 25 words. No jargon.
Format: Return ONLY the 3 bullets, nothing else."""

print('Bad prompt 1:', bad_prompts[0])
print()
print('Good rewrite:')
print(good_example)
print()
print('Response:')
print(ask(good_example))
```

## Part 2: Role Prompting

```python
question = 'How does a computer store information?'

roles = [
    'You are a kindergarten teacher explaining to a 5-year-old.',
    'You are a computer science professor teaching graduate students.',
    'You are a stand-up comedian making a joke about technology.',
]

for role in roles:
    print(f'ROLE: {role[:60]}...')
    print(ask(question, system=role, max_tokens=150))
    print('-' * 50)
```

## Part 3: Few-Shot Examples

```python
zero_shot = 'Classify the sentiment of this review: "The battery dies after 2 hours. Terrible product."'

few_shot = """Classify sentiment as POSITIVE, NEGATIVE, or NEUTRAL.

Examples:
Review: "Best phone I've ever owned!" → POSITIVE
Review: "It arrived damaged." → NEGATIVE  
Review: "It's a phone. Does phone things." → NEUTRAL

Review: "The battery dies after 2 hours. Terrible product." → """

print('Zero-shot response:')
print(ask(zero_shot))
print()
print('Few-shot response:')
print(ask(few_shot))
```

## Part 4: Chain-of-Thought

```python
problem = 'A store sells apples for $0.50 each or 6 for $2.50. How much do I save buying 12 apples in bulk vs individually?'

no_cot = ask(problem)
cot_prompt = problem + '\n\nThink through this step by step before giving your final answer.'
cot = ask(cot_prompt, max_tokens=400)

print('WITHOUT chain-of-thought:')
print(no_cot)
print()
print('WITH chain-of-thought:')
print(cot)
```

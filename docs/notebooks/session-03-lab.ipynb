## Part 1: List and Dict Manipulation

```python
# AI chat history is a list of dicts — practice this pattern
chat_history = []

def add_message(role, content):
    chat_history.append({'role': role, 'content': content})

add_message('user', 'What is machine learning?')
add_message('assistant', 'ML is a way for computers to learn from data.')
add_message('user', 'Can you give an example?')

print(f'Total messages: {len(chat_history)}')
for msg in chat_history:
    print(f"  [{msg['role']:10s}] {msg['content']}")
```

## Part 2: File I/O — Read and Summarize a CSV

```python
import csv
import io

# Simulate a CSV of AI model data
csv_data = '''model,company,params_B,release_year,context_window_k
GPT-4,OpenAI,1800,2023,128
Claude 3 Opus,Anthropic,,,200
Gemini Ultra,Google,,,32
Llama 3.1 405B,Meta,405,2024,128
Mistral Large,Mistral AI,123,2024,32
'''

reader = csv.DictReader(io.StringIO(csv_data))
models = list(reader)

print(f'Loaded {len(models)} models')
print(f'Columns: {list(models[0].keys())}')
print()
for m in models:
    print(f"  {m['model']:25s} by {m['company']:15s} ({m['release_year'] or 'N/A'})")
```

## Part 3: Functions and Error Handling

```python
import anthropic
import os

os.environ.setdefault('ANTHROPIC_API_KEY', 'your-key-here')
client = anthropic.Anthropic()

def summarize_csv(csv_text: str, question: str) -> str:
    """Send a CSV to Claude and ask a question about it."""
    try:
        resp = client.messages.create(
            model='claude-haiku-4-5-20251001',
            max_tokens=300,
            messages=[{'role': 'user', 'content': f'CSV data:\n{csv_text}\n\nQuestion: {question}'}]
        )
        return resp.content[0].text
    except Exception as e:
        return f'Error: {e}'

result = summarize_csv(csv_data, 'Which model has the largest context window?')
print(result)
```

## Part 4: Build a Simple Chat Loop

```python
def chat_once(messages: list, user_input: str) -> tuple[str, list]:
    """Send one message, return reply and updated history."""
    messages = messages + [{'role': 'user', 'content': user_input}]
    resp = client.messages.create(
        model='claude-haiku-4-5-20251001',
        max_tokens=200,
        messages=messages
    )
    reply = resp.content[0].text
    messages = messages + [{'role': 'assistant', 'content': reply}]
    return reply, messages

# Test it
history = []
questions = ['What is Python?', 'What is it good for in AI?']

for q in questions:
    print(f'You: {q}')
    reply, history = chat_once(history, q)
    print(f'AI:  {reply[:200]}')
    print()
```

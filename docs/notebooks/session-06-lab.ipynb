```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

np.random.seed(42)
X, y = make_regression(n_samples=100, n_features=1, noise=15)
X = X.flatten()

plt.figure(figsize=(8, 4))
plt.scatter(X, y, alpha=0.6, color='#60a5fa')
plt.title('Training Data')
plt.xlabel('Feature (X)')
plt.ylabel('Target (y)')
plt.grid(True, alpha=0.3)
plt.show()
```

## Part 1: Gradient Descent from Scratch

```python
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def train_linear(X, y, lr=0.01, epochs=100):
    w = 0.0  # weight
    b = 0.0  # bias
    n = len(X)
    losses = []

    for epoch in range(epochs):
        y_pred = w * X + b
        loss = mse_loss(y, y_pred)
        losses.append(loss)

        # Gradients
        dw = -2/n * np.sum(X * (y - y_pred))
        db = -2/n * np.sum(y - y_pred)

        # Update
        w -= lr * dw
        b -= lr * db

        if epoch % 20 == 0:
            print(f'Epoch {epoch:3d}: loss={loss:.2f}, w={w:.3f}, b={b:.3f}')

    return w, b, losses

w, b, losses = train_linear(X, y, lr=0.1, epochs=100)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Loss curve
ax1.plot(losses, color='#f87171')
ax1.set_title('Training Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('MSE Loss')
ax1.grid(True, alpha=0.3)

# Fitted line
ax2.scatter(X, y, alpha=0.6, color='#60a5fa', label='Data')
X_line = np.linspace(X.min(), X.max(), 100)
ax2.plot(X_line, w * X_line + b, color='#f87171', linewidth=2, label=f'y={w:.2f}x+{b:.2f}')
ax2.set_title('Fitted Line')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

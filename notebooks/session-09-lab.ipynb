```python
import tiktoken

enc = tiktoken.encoding_for_model('gpt-4')

samples = [
    ('Tweet', 'AI is transforming every industry! #MachineLearning'),
    ('Python code', 'def fibonacci(n):\n    if n <= 1: return n\n    return fibonacci(n-1) + fibonacci(n-2)'),
    ('Japanese', '人工知能は世界を変えています'),
    ('Paragraph', 'Machine learning is a subset of artificial intelligence that enables computers to learn from data without being explicitly programmed. It uses statistical techniques to give computer systems the ability to progressively improve performance on specific tasks.')
]

print(f'{"Type":<15} {"Tokens":>8} {"Chars":>8} {"Ratio":>8}')
print('-' * 45)
for name, text in samples:
    tokens = enc.encode(text)
    print(f'{name:<15} {len(tokens):>8} {len(text):>8} {len(text)/len(tokens):>8.2f}')
```

```python
import anthropic, os
os.environ.setdefault('ANTHROPIC_API_KEY', 'your-key-here')
client = anthropic.Anthropic()

# Demonstrate context window: with vs without document
document = 'The Zephyr Initiative was founded in 2019 by Dr. Maya Thornton. Its annual budget is $3.7M.'
question = 'What is the annual budget of the Zephyr Initiative?'

with_ctx = client.messages.create(
    model='claude-haiku-4-5-20251001', max_tokens=100,
    messages=[{'role': 'user', 'content': f'Document: {document}\n\nQuestion: {question}'}]
).content[0].text

without_ctx = client.messages.create(
    model='claude-haiku-4-5-20251001', max_tokens=100,
    messages=[{'role': 'user', 'content': question}]
).content[0].text

print('WITH document:', with_ctx)
print()
print('WITHOUT document (hallucination risk):', without_ctx)
```
